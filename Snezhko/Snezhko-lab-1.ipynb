# II. Проверить данные на нормальность, используя критерии:
#### 1. Хи-квадрат (Реализовать самому, не используя библиотечные варианты!!)

```python
import numpy as np
from scipy.stats import norm, chi2

def chi_square_normality_test(data, alpha=0.03):
    """
    Проверка нормальности распределения выборки с помощью критерия χ².
    
    Parameters:
        data : list или np.array — данные
        alpha : уровень значимости
        
    Returns:
        dict с результатами:
            chi2_stat : значение статистики χ²
            df        : число степеней свободы
            p_value   : p-значение
            is_normal : True/False — гипотеза H0 о нормальности
    """
    data = np.asarray(data)
    n = len(data)

    # Число интервалов (правило Стерджеса)
    k = 1 + int(np.log2(n))

    # Разбиение на интервалы
    intervals = np.linspace(min(data), max(data), k + 1)

    # Реальные частоты
    real_freq, bin_edges = np.histogram(data, bins=intervals)

    # Математическое ожидание и дисперсия по выборке
    mu = np.mean(data)
    sigma = np.std(data, ddof=1)

    # Ожидаемые частоты
    expected_freq = np.array([
        norm.cdf(bin_edges[i+1], mu, sigma) - norm.cdf(bin_edges[i], mu, sigma)
        for i in range(k)
    ]) * n

    # Статистика χ²
    chi2_stat = np.sum((real_freq - expected_freq) ** 2 / (expected_freq + 1e-9))

    # Степени свободы: (число интервалов - 1 - число оценённых параметров mu и sigma)
    df = k - 1 - 2

    # p-значение
    p_value = 1 - chi2.cdf(chi2_stat, df)

    # Решение: нормальны ли данные
    is_normal = p_value > alpha

    return {
        "chi2_stat": chi2_stat,
        "df": df,
        "p_value": p_value,
        "is_normal": is_normal,
        "n_bins_used": k
    }
```

```python
chi_square_normality_test(data[col])
```

# III. Используя техники обработки данных, постарайтесь привести данные к нормальному распределению

#### 1. Удаление, усечение выбросов

```python
from collections import Counter

sorted(Counter(data[col]).keys())
```

```python
# IQR
q1 = series.quantile(0.25)
q3 = series.quantile(0.75)
iqr = q3 - q1

# Границы
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

series_trimmed = series[(series >= lower) & (series <= upper)]
print(f"Удалено выбросов: {len(series) - len(series_trimmed)}")
```

Явных выбросов нет

#### 2. Стандартизация, нормировка

```python
series_standardized = (series_trimmed - series_trimmed.mean()) / series_trimmed.std(ddof=0)
```

```python
series_normed = (series_trimmed - series_trimmed.min()) / (series_trimmed.max() - series_trimmed.min())
```

#### 3. Логарифмирование и т.п.

```python
if (series_trimmed <= 0).any():
    shift = abs(series_trimmed.min()) + 1e-6
else:
    shift = 0
series_log = np.log(series_trimmed + shift)
```

Удаление выбросов сглаживает хвосты, уменьшает эксцесс.

Логарифмирование уменьшает правую асимметрию.

Стандартизация и нормировка полезны для ML, но не сильно меняют форму распределения.

```python
variants = {
    "Original": series,
    "Trimmed": series_trimmed,
    "Standardized": series_standardized,
    "Normalized": series_normed,
    "Log-transformed": series_log
}

for name, s in variants.items():
    plt.figure(figsize=(6,4))
    plt.hist(s, bins=30, edgecolor="black", alpha=0.7)
    plt.title(f"Гистограмма: {name}")
    plt.xlabel("Значение")
    plt.ylabel("Частота")
    plt.show()
```

# IV. Сгруппируйте данные по столбцу 'School_Grade'. 

#### 1. На одном графике постройте гистограммы для каждой группы.

```python
# Группировка по School_Grade
grouped = data.groupby("School_Grade")[col]

plt.figure(figsize=(10,6))
bins = np.linspace(data[col].min(), data[col].max(), 15)

for grade, group in grouped:
    plt.hist(group.dropna(), bins=bins, alpha=0.5, label=str(grade), edgecolor="black")

plt.title(f"Гистограммы {col} по School_Grade")
plt.xlabel(col)
plt.ylabel("Частота")
plt.legend(title="School_Grade")
plt.show()
```

#### 2. Рассчитайте среднее, дисперсию внутри каждой группы

```python
stats_table = grouped.agg(['mean', 'var']).reset_index()
print(stats_table)
```

Существуют небольшие различия распределений между классами. Старшие классы демонстрируют более стабильное поведение, младшие — более вариативное.
