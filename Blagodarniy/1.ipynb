## I. 

```python
data = df[my_column]

mean_val = data.mean()
variance_val = data.var(ddof=1)
mode_val = data.mode().iloc[0]
median_val = data.median()
quantiles_val = data.quantile([0.25, 0.5, 0.75])
kurtosis_val = stats.kurtosis(data, fisher=True)
skewness_val = stats.skew(data)
iqr_val = stats.iqr(data)

print(f"Среднее: {mean_val}")
print(f"Дисперсия: {variance_val}")
print(f"Мода: {mode_val}")
print(f"Медиана: {median_val}")
print(f"Квантили:\n{quantiles_val}")
print(f"Эксцесс: {kurtosis_val}")
print(f"Асимметрия: {skewness_val}")
print(f"IQR: {iqr_val}")
```

## II. 


```python
data = df[my_column].dropna()
n = len(data)
bins = int(np.sqrt(n))


hist, bin_edges = np.histogram(data, bins=bins)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
mean_val = np.mean(data)
std_val = np.std(data, ddof=1)

expected_probs = [stats.norm.cdf(bin_edges[i+1], mean_val, std_val) - 
                  stats.norm.cdf(bin_edges[i], mean_val, std_val) for i in range(bins)]
expected_counts = np.array(expected_probs) * n

chi2_stat = np.sum((hist - expected_counts)**2 / expected_counts)
df_chi2 = bins - 3
p_value = 1 - stats.chi2.cdf(chi2_stat, df_chi2)

print(f"Chi-square statistic: {chi2_stat:.2f}, p-value: {p_value:.4f}")
```

```python
skewness_val = stats.skew(data)
kurtosis_val = stats.kurtosis(data) 
print(f"Асимметрия: {skewness_val:.4f}")
print(f"Эксцесс: {kurtosis_val:.4f}")
```

```python
stats.probplot(data, dist="norm", plot=plt)
plt.title("Q-Q plot")
plt.show()
```

### Нормальность данных

Chi-square statistic: 2356.47, p-value = 0 → данные не нормальны.

Асимметрия и эксцесс также указывают на отклонение от нормального распределения.

Q-Q plot показывает, что хвосты распределения отклоняются от прямой.

## III.

```python
# удаление выбросов
Q1 = df[my_column].quantile(0.25)
Q3 = df[my_column].quantile(0.75)

lower = Q1 - 1.5 * iqr_val
upper = Q3 + 1.5 * iqr_val

df_clean = df[my_column][(df[my_column] >= lower) & (df[my_column] <= upper)].to_frame()
```

```python
# стандартизация
mean = df_clean[my_column].mean()
std = df_clean[my_column].std()
df_clean['standardized'] = (df_clean[my_column] - mean) / std
```

```python
# логарифмирование
df_clean['log_transformed'] = np.log1p(df_clean[my_column]) # log(x+1), чтобы избежать log(0)
```

```python
# стандартизируем
df_clean['full_standardized'] = (df_clean['log_transformed'] - df_clean['log_transformed'].mean()) / df_clean['log_transformed'].std()
```

```python
df_clean
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten() 

columns = ['Time_on_Gaming',  'standardized', 'log_transformed', 'full_standardized']

for i, col in enumerate(columns):
    axes[i].hist(df_clean[col], bins=30, edgecolor='black')
    axes[i].set_title(f'Гистограмма {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Частота')

plt.tight_layout()
plt.show()
```

### Эффект обработки данных

После удаления выбросов хвосты уменьшились.

Стандартизация привела к среднему ≈ 0 и дисперсии ≈ 1, сохранив форму распределения.

Логарифмирование (log1p) смягчило правый хвост, распределение стало ближе к нормальному, но всё ещё есть лёгкая асимметрия.

Полностью нормализовать данные без сильного искажения невозможно, но логарифмирование и стандартизация значительно приближают распределение к нормальному виду.
