### Тривиальный классификатор

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix

def eval_model(model, X_test, y_test, write=True):
    y_pred = model.predict(X_test)
    metrics = {
        'acc': accuracy_score(y_test, y_pred),
        'prec': precision_score(y_test, y_pred, zero_division=0),
        'rec': recall_score(y_test, y_pred, zero_division=0)
    }

    metrics['f1'] = 2 * metrics['prec'] * metrics['rec'] / (metrics['prec'] + metrics['rec']) if metrics['prec'] + metrics['rec'] > 0 else 0
    
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)
    else:
        y_proba = y_pred

    metrics['auc'] = roc_auc_score(y_test, y_proba)

    if (write): 
        for metric, val in metrics.items():
            print(f'{metric}:{val}')
    
    return metrics
```

```python
from sklearn.base import RegressorMixin
class BaseLineRegressor(RegressorMixin):
    def fit(self, X, y):
        values, counts = np.unique(y, return_counts=True)
        self.most_common  = values[counts.argmax()]
        return self

    def predict(self, X):
        return np.full(X.shape[0], self.most_common)
```

### Байесовский классификатор

```python
class GaussianNaiveBayes(RegressorMixin):
    def __init__(self, var_smoothing=1e-9):
        self.var_smoothing = var_smoothing
        
    def fit(self, X, y, numeric_cols):
        self.numeric_cols=numeric_cols
        self.classes_ = np.unique(y)
        self.priors_ = []
        self.means_ = []
        self.vars_ = []
        self.p = []

        X_num = X[:, self.numeric_cols]
        
        for cls in self.classes_:
            X_cls = X_num[y == cls]
            self.priors_.append(len(X_cls) / len(X))
            self.means_.append(X_cls.mean(axis=0))
            self.vars_.append(X_cls.var(axis=0) + self.var_smoothing)
        
        self.priors_ = np.array(self.priors_)
        self.means_ = np.array(self.means_)
        self.vars_ = np.array(self.vars_)

        for i in range(X.shape[1]):
            if i in self.numeric_cols:
                self.p.append({})
                continue

            d = dict()

            for val in np.unique(X[:, i]):
                mask = (X[:, i] == val)
                y1 = y[mask]
                X1 = X[mask]
                from collections import Counter
                cnts = dict(Counter(y1))
                for j in self.classes_:
                    if j not in cnts.keys():
                        cnts[j] = 0
                d[val] = [cnt / len(X1) for _, cnt in sorted(cnts.items())]
        
            self.p.append(d)
            
        return self
    
    def _log_likelihood(self, X):
        n_samples = X.shape[0]
        n_classes = len(self.classes_)
        log_likelihood = np.zeros((n_samples, n_classes))
        
        for i in range(n_classes):
            diff = X - self.means_[i]
            log_likelihood[:, i] = -0.5 * np.sum(
                np.log(2 * np.pi * self.vars_[i]) + (diff ** 2) / self.vars_[i],
                axis=1
            )
        
        return log_likelihood
    
    def predict_proba(self, X):
        X_num = X[:, self.numeric_cols]
        log_likelihood = self._log_likelihood(X_num)
        log_posterior = log_likelihood + np.log(self.priors_)
        max_log = np.max(log_posterior, axis=1, keepdims=True)
        exp_log = np.exp(log_posterior - max_log)
        probas =  exp_log / np.sum(exp_log, axis=1, keepdims=True)

        for i in range(X.shape[1]):
            if i in self.numeric_cols:
                continue

            p = np.array(list(self.p[i].get(x, np.zeros(len(self.classes_))) for x in X[:, i]))

            probas *= p
        
        return probas[:, 1].reshape(-1)
    
    def predict(self, X):
        proba = self.predict_proba(X)
        return np.round(proba)
```

### Метод k-ближайших соседей

```python
import numpy as np
from collections import Counter
from scipy.spatial import distance
from sklearn.preprocessing import StandardScaler

class KNN:
    def __init__(self, k=3, metric='euclidean'):
        self.k = k
        self.metric = metric
        
    def fit(self, X, y):
        self.scaler = StandardScaler()
        self.scaler.fit(X, y)
        X = self.scaler.transform(X)
        self.X_train = np.array(X)
        self.y_train = np.array(y)
        return self
    
    def _distance(self, a, b):
        if self.metric == 'euclidean':
            return np.sqrt(np.sum((a - b) ** 2))
        elif self.metric == 'manhattan':
            return np.sum(np.abs(a - b))
        else:
            return distance.cdist([a], [b], self.metric)[0][0]
    
    def predict(self, X):
        X = np.array(X)
        X = self.scaler.transform(X)
        predictions = []
        
        for x in X:
            distances = [self._distance(x, x_train) for x_train in self.X_train]
            k_indices = np.argsort(distances)[:self.k]
            k_labels = self.y_train[k_indices]
            most_common = Counter(k_labels).most_common(1)[0][0]
            predictions.append(most_common)
        
        return np.array(predictions)
    
    def predict_proba(self, X):
        X = np.array(X)
        X = self.scaler.transform(X)
        probas = []
        
        for x in X:
            distances = [self._distance(x, x_train) for x_train in self.X_train]
            k_indices = np.argsort(distances)[:self.k]
            k_labels = self.y_train[k_indices]
            
            unique_classes = np.unique(self.y_train)
            proba = [np.sum(k_labels == c) / self.k for c in unique_classes]
            probas.append(proba)
        
        return np.array(probas)[:, 1].reshape(-1)
```

### Логистическая регрессия

```python
from scipy.special import expit
np.random.seed(42)

class LogisticRegression(RegressorMixin):

    def __init__(self, alpha=0, lr=0.5, max_iter=10000,
                 fit_intercept=True):
        self.alpha = alpha
        self.lr = lr
        self.max_iter = int(max_iter)
        self.fit_intercept = fit_intercept
        self.batch_size = 512

    @staticmethod
    def _sigmoid(x):
        return expit(x)

    def _add_intercept(self, X):

        X_copy = np.full((X.shape[0], X.shape[1] + 1), fill_value=1.)
        X_copy[:, :-1] = X

        return X_copy

    def fit(self, X, Y):
        self.scaler = StandardScaler()
        self.scaler.fit(X, Y)
        X = self.scaler.transform(X)
        assert X.shape[0] == Y.shape[0]

        if self.fit_intercept:
            X_copy = self._add_intercept(X)
        else:
            X_copy = X.copy()

        W = np.zeros((X_copy.shape[1], 1))

        for _ in range(0, self.max_iter):
            batch = np.random.choice(X_copy.shape[0], size = min(self.batch_size, X_copy.shape[0]), replace=False)
            x = X_copy[batch]
            y = Y[batch].reshape((len(batch), 1))

            grad = (x.T @ (self._sigmoid(x @ W) - y))
            reg = self.alpha * W
            if self.fit_intercept:
                reg[-1] = 0
            grad += reg

            W -= self.lr * grad

        if self.fit_intercept:
            self.coef_ = W[:-1]
            self.intercept_ = W[-1][0]
        else:
            self.coef_ = W[:]
            self.intercept_ = 0

        self.weights = W
        return self

    def predict(self, X):
        X = self.scaler.transform(X)
        if self.fit_intercept:
            X_copy = self._add_intercept(X)
        else:
            X_copy = X.copy()

        assert X_copy.shape[1] == self.weights.shape[0]

        predictions = np.round(self._sigmoid(X_copy @ self.weights)).reshape(X.shape[0])

        return predictions

    def predict_proba(self, X):
        X = self.scaler.transform(X)
        if self.fit_intercept:
            X_copy = self._add_intercept(X)
        else:
            X_copy = X.copy()

        assert X_copy.shape[1] == self.weights.shape[0]

        prob_predictions = self._sigmoid(X_copy @ self.weights).reshape(X.shape[0])

        return prob_predictions
```

```python
f1s = []
for alpha in [0, 0.1, 0.3, 0.5, 0.7, 1]:
    for lr in [1, 0.5, 0.01, 0.001, 0.0001, 0.00001]:
        model = LogisticRegression(alpha=alpha, lr=lr).fit(X_train, y_train)
        res = eval_model(model, X_val, y_val, write=False)
        f1s.append([res["f1"], alpha, lr])

f1s = np.array(f1s)
i = np.argmax(f1s[:, 0])

best_f1 = f1s[i, 0]
best_alpha = f1s[i, 1]
best_lr = f1s[i, 2]
print(best_f1, best_alpha, best_lr)
```
